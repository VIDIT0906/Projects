```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
weather = pd.read_csv('Test.csv')
```

```python
weather.head()
```

```python
weather.info()
```

```python
sns.barplot(x = weather['humidity'], 
            y = weather['temperature'])
```

```python
sns.displot(weather['humidity'])
```

```python
sns.jointplot(x = weather['humidity'], 
              y = weather['temperature'])
```

```python
sns.jointplot(x = weather['humidity'], 
              y = weather['temperature'], kind="hex")
```

```python
sns.stripplot(x = weather['weather_type'], 
              y = weather['temperature'])
```

```python
sns.boxplot(x = weather['humidity'], 
            y = weather['temperature'], 
            hue=weather['weather_type'])
```

```python
sns.barplot(x = weather['humidity'], 
            y = weather['temperature'], 
            hue=weather['weather_type'])
```

```python
sns.countplot(x = weather['weather_type'])
```

```python
sns.pointplot(x = weather['humidity'], 
              y = weather['temperature'], 
              hue=weather['weather_type'])
```

```python
sns.lmplot(x="humidity", y="temperature", data=weather)
```
